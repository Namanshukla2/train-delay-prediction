```python
import pandas as pd
import numpy as np

print("Pandas version:", pd.__version__)
```

```python
file_path = "../data/raw/indian_delay/combined_delay.csv"

df = pd.read_csv(file_path)

print("Total rows:", len(df))
df.head()
```

```python
# ================================
# Basic Dataset Information
# ================================

# Display shape
print("Initial Dataset Shape:", df.shape)

# Check data types
print("\nData Types:")
print(df.dtypes)

# Check missing values
print("\nMissing Values:")
print(df.isna().sum())
```

```python
# ================================
# Convert 'date' column to datetime
# ================================

df['date'] = pd.to_datetime(df['date'])

# Extract useful time-based features
df['year'] = df['date'].dt.year
df['month'] = df['date'].dt.month
df['day'] = df['date'].dt.day
df['day_of_week'] = df['date'].dt.dayofweek  # Monday=0, Sunday=6

print("Date conversion complete ✅")
df.head()
```

```python
# ================================
# Remove rows where delay is missing
# ================================

df_clean = df.dropna(subset=['delay'])

print("Rows after removing missing delay:", len(df_clean))
```

```python
# ================================
# Remove unrealistic delay values
# Keep delay between -180 and 720 minutes
# ================================

df_clean = df_clean[
    (df_clean['delay'] >= -180) &
    (df_clean['delay'] <= 720)
]

print("Rows after outlier removal:", len(df_clean))

# Check new delay statistics
print("\nCleaned Delay Statistics:")
print(df_clean['delay'].describe())
```

```python
# ==========================================
# Create Training Sample (20% of data)
# ==========================================

# Set random seed for reproducibility
RANDOM_STATE = 42

# Take 20% random sample
df_sample = df_clean.sample(frac=0.2, random_state=RANDOM_STATE)

print("Sampled Dataset Shape:", df_sample.shape)
```

```python
import os
os.makedirs("../data/processed", exist_ok=True)
```

```python
# ==========================================
# Save cleaned & sampled dataset
# ==========================================

output_path = "../data/processed/cleaned_sampled_delay.csv"

df_sample.to_csv(output_path, index=False)

print("✅ Cleaned and sampled dataset saved successfully!")
```

```python
# ==========================================
# Clean combined_schedule.csv
# ==========================================

schedule_df = pd.read_csv("../data/raw/indian_delay/combined_schedule.csv")

# Convert arrival & departure time to datetime format
schedule_df['arrival_time'] = pd.to_datetime(
    schedule_df['arrival_time'], format='%H:%M', errors='coerce'
)

schedule_df['departure_time'] = pd.to_datetime(
    schedule_df['departure_time'], format='%H:%M', errors='coerce'
)

print("✅ Schedule time columns converted")
schedule_df.head()
```

```python
# ==========================================
# Create Route Distance & Stop Count
# ==========================================

route_features = schedule_df.groupby('train_no').agg({
    'distance_from_origin': 'max',
    'station_no': 'count'
}).reset_index()

route_features.columns = [
    'train_no',
    'total_route_distance',
    'total_stops'
]

print("✅ Route features created")
route_features.head()
```

```python
# ==========================================
# Merge Route-Level Features into Delay Dataset
# ==========================================

# Load cleaned sampled dataset
delay_df = pd.read_csv("../data/processed/cleaned_sampled_delay.csv")

# Merge route features
delay_df = delay_df.merge(route_features, on='train_no', how='left')

print("✅ Route features merged successfully")
print("New shape:", delay_df.shape)

delay_df.head()
```

```python
# ==========================================
# Merge Train Type Information
# ==========================================

train_df = pd.read_csv("../data/raw/indian_delay/train_details.csv")

# Merge on train_no
delay_df = delay_df.merge(train_df[['train_no', 'type_code']],
                          on='train_no',
                          how='left')

print("✅ Train type merged successfully")
print("New shape:", delay_df.shape)

# Check missing values in type_code
print("\nMissing train types:", delay_df['type_code'].isna().sum())

delay_df.head()
```

```python
# ==========================================
# Merge Station Zone Information
# ==========================================

station_df = pd.read_csv("../data/raw/indian_delay/station_full_names.csv")

# Merge on station_name
delay_df = delay_df.merge(
    station_df[['station_name', 'station_zone']],
    on='station_name',
    how='left'
)

print("✅ Station zone merged successfully")
print("New shape:", delay_df.shape)

# Check missing zones
print("\nMissing station_zone:", delay_df['station_zone'].isna().sum())

delay_df.head()
```

```python
# ==========================================
# Check for duplicates
# ==========================================

duplicate_count = delay_df.duplicated().sum()
print("Duplicate rows:", duplicate_count)

# Check null values overall
print("\nMissing values per column:")
print(delay_df.isna().sum())
```

```python
# ==========================================
# Clean categorical columns
# ==========================================

delay_df['type_code'] = delay_df['type_code'].str.strip()
delay_df['station_zone'] = delay_df['station_zone'].str.strip()

print("✅ Categorical cleaning done")
```

```python
# ==========================================
# Handle Missing Values
# ==========================================

# Fill numeric missing with median
delay_df['total_route_distance'].fillna(
    delay_df['total_route_distance'].median(),
    inplace=True
)

delay_df['total_stops'].fillna(
    delay_df['total_stops'].median(),
    inplace=True
)

# Fill categorical missing with 'Unknown'
delay_df['station_zone'].fillna('Unknown', inplace=True)

print("✅ Missing values handled successfully")

print("\nRemaining missing values:")
print(delay_df.isna().sum())
```

```python
# ==========================================
# Proper Missing Value Handling (Safe Way)
# ==========================================

# Fill numeric columns
delay_df['total_route_distance'] = delay_df['total_route_distance'].fillna(
    delay_df['total_route_distance'].median()
)

delay_df['total_stops'] = delay_df['total_stops'].fillna(
    delay_df['total_stops'].median()
)

# Fill categorical column
delay_df['station_zone'] = delay_df['station_zone'].fillna('Unknown')

print("✅ Missing values handled correctly")

print("\nRemaining missing values:")
print(delay_df.isna().sum())
```

```python
# ==========================================
# Save Final Cleaned Dataset
# ==========================================

import os

# Ensure processed folder exists
os.makedirs("../data/processed", exist_ok=True)

# Save final dataset
final_path = "../data/processed/final_cleaned_dataset.csv"
delay_df.to_csv(final_path, index=False)

print("✅ Final cleaned dataset saved successfully!")
print("Path:", final_path)
```
